### Original SVGD

```python
@jax.jit
def gaussian(x, mu, sigma):
    return np.exp(- 0.5 * (x - mu) ** 2 / sigma ** 2)

@jax.jit
def log_gaussian(x, mu, sigma):
    return np.log(gaussian(x, mu, sigma))

@jax.jit
def log_gaussian_dx(x, mu, sigma):
    return jax.vmap(jax.grad(log_gaussian), (0, None, None))(x, mu, sigma)


# ORIGINAL SVGD IMPLEMENTATION

@jax.jit
def SVGD_kernel(theta, h):
    '''
    Function taken from: https://github.com/dilinwang820/Stein-Variational-Gradient-Descent/blob/master/python/svgd.py
    '''
    # sq_dist = pdist(theta)
    # pairwise_dists = squareform(sq_dist) ** 2
    theta_norm_squared = np.sum(theta ** 2, axis=1, keepdims=True)
    pairwise_dists = theta_norm_squared + theta_norm_squared.T - 2 * np.dot(theta, theta.T)
    # if h < 0: # median trick
    #     h = jnp.median(pairwise_dists)  
    #     h = jnp.sqrt(0.5 * h / jnp.log(theta.shape[0] + 1))

    # compute the rbf kernel
    Kxy = np.exp(- pairwise_dists / h ** 2 / 2)

    dxkxy = - np.matmul(Kxy, theta)
    sumkxy = np.sum(Kxy, axis=1)
    # for i in range(theta.shape[1]):
    #     dxkxy = dxkxy.at[:, i].set(dxkxy[:, i] + jnp.multiply(theta[:, i], sumkxy))
    dxkxy += np.multiply(theta, np.expand_dims(sumkxy, axis=1)) # vectorized
    dxkxy /= (h ** 2)
    return (Kxy, dxkxy)


def SVGD_update(x0, lnprob, n_iter=1000, stepsize=1e-3, alpha=1.0, debug=False):
    '''
    Function taken from: https://github.com/dilinwang820/Stein-Variational-Gradient-Descent/blob/master/python/svgd.py
    '''
    theta = np.copy(x0)

    for iter in range(n_iter):
        
        lnpgrad = lnprob(theta.squeeze()).reshape(-1, 1) # lnpgrad: (n, d)
        # calculating the kernel matrix
        kxy, dxkxy = SVGD_kernel(theta, h=0.05) # kxy: (n, n), dxkxy: (n, d)
        grad_theta = alpha * (np.matmul(kxy, lnpgrad) + dxkxy) / x0.shape[0] # grad_theta: (n, d)
        
        # vanilla update
        theta = theta + stepsize * grad_theta

    plt.plot(kxy[0])
    plt.show()
        
    return theta
```

```python
dlnprob = lambda x: log_gaussian_dx(x, 0, 1)
x = SVGD_update(x0.reshape(-1, 1), dlnprob, n_iter=1000, stepsize=0.01)
```

```python
dlnprob = lambda x: log_gaussian_dx(x, 0, 1)
start = time.time()
for i in range(20):
    x = SVGD_update(x0.reshape(-1, 1), dlnprob, n_iter=1000, stepsize=0.01)
print('Average time:', (time.time() - start) / 20)
```

### SVGD

```python
h = 1e-6 # must be very small for Cholesky (otherwise NaNs)

def K(x):
    xmx = np.expand_dims(x, 0) - np.expand_dims(x, 1)
    norm = np.einsum('ijk,ijk->ij', xmx, xmx)
    return np.exp(- (norm) / h)

def g_K(x):
    # we avoid calling autograd since the function is non-scalar, for better efficiency
    xmx = np.expand_dims(x, 0) - np.expand_dims(x, 1)
    return np.expand_dims(K(x), -1) * (2. * xmx / h)

def logp(x):
    # Standard Gaussian target
    return -np.sum(x**2)

def p(x):
    return np.exp(-x**2)
```

```python
x_test = np.linspace(-1, 1, 100)
plt.plot(x_test, K(x)[0])
# plt.plot(x_test, g_K(x))
plt.show()
```

```python
def svgd(x0, logp, T=100, eta=0.1):
    
    x = x0
    g_logp = grad(logp)

    update_svgd = jit(lambda x: K(x) @ g_logp(x) + g_K(x).sum(0))

    for i in range(T):
        x = x + eta * update_svgd(x)
        if np.any(np.isnan(x)):
            print(f'NaNs at iteration {i}')
            break
    
    return x
```

```python
start = time.time()
for i in range(20):
    x = svgd(x0, logp, T=1000, eta=0.01)
print('Average time:', (time.time() - start) / 20)
```

### SGLD-R

```python
def sgldr(x0, logp, T=100, eta=0.1, key=key):
    
    x = x0
    g_logp = grad(logp)
    xs = []

    update_sgldr = jit( lambda x, key : ( eta * (K(x) @ g_logp(x) + g_K(x).sum(0)) + \
                                         np.linalg.cholesky(2 * eta * K(x)) @ random.normal(key, (N, d)), random.split(key)[0] ) )

    for i in range(T):
        x_upd, key = update_sgldr(x, key)
        x = x + x_upd
        xs.append(x)
        if np.any(np.isnan(x)):
            print(f'NaNs at iteration {i}')
            break
    
    return x, xs
```

```python
x, xs = sgldr(x0, logp, T=1000, eta=0.01, key=key)
```

# Wasserstein distance for convergence monitoring

```python
N = 100

key, subkey = random.split(random.PRNGKey(1))
x0 = 3. + 0.5 * random.normal(key, (N, d))
```

```python
h = 1e-1 # must be very small for Cholesky (otherwise NaNs)

def K(x):
    xmx = np.expand_dims(x, 0) - np.expand_dims(x, 1) # pairwise differences
    norm = np.einsum('ijk,ijk->ij', xmx, xmx) # returns a 2D matrix of element-wise products
    return np.exp(- (norm) / h**2)

def g_K(x):
    # we avoid calling autograd since the function is non-scalar, for better efficiency
    xmx = np.expand_dims(x, 0) - np.expand_dims(x, 1)
    return np.expand_dims(K(x), -1) * (2. * xmx / h**2)

def logp(x):
    # Standard Gaussian target
    return -np.sum(x**2)

def p(x):
    return np.exp(-x**2)
```

```python
def svgd2(x0, logp, T=100, eta=0.1, tol=1e-3, p=1):
    
    x = np.copy(x0)
    x_old = np.copy(x0)
    g_logp = grad(logp)
    wass = []

    update_svgd = jit(lambda x: K(x) @ g_logp(x) + g_K(x).sum(0)) 

    for i in range(T):
        x = x + eta * update_svgd(x)
        w = wasserstein_1d(x_old.squeeze(), x.squeeze(), p=p)
        wass.append(w)
        x_old = np.copy(x)
    
    return x, wass
```

```python
x, wass = svgd2(x0, logp, T=1000, eta=0.01, p=2)
```

```python
def sgd(x0, logp, T=100):
    
    x = x0
    g_logp = grad(logp)

    update_sgd = jit( lambda x : g_logp(x) ) 

    for i in range(T):
        x = x + eta * update_sgd(x)   
    return x
```

```python
def sgdm(x0, logp, T=100, alpha=0.5):
    
    x = x0
    g_logp = grad(logp)

    update_sgd = jit( lambda x : g_logp(x) ) 
    m = np.zeros_like(x)

    for i in range(T):
        m = alpha * m + eta * update_sgd(x)   
        x = x + m
    return x
```

```python
def sgdmr(x0, logp, T=100, alpha=0.9):
    
    x = x0
    g_logp = grad(logp)

    update_sgdm1 = jit( lambda x, m : K(x) @ m + g_K(x).sum(0) ) 
    update_sgdm2 = jit( lambda x : K(x) @ g_logp(x) + g_K(x).sum(0) ) 
    m = np.zeros_like(x0)

    for i in range(T):
        x = x + eta * update_sgdm1(x, m)
        m = alpha * m + eta * update_sgdm2(x)
    return x
```
